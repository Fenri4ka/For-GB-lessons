f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from math import *
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
func = -12 * x ** 4 * sin(cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
solve(func, x)
```

Не удаётся найти конкретные корни уравнения, т.к. их бесконечное количество (см. график)

```python
solve(diff(func) > 0)
```

```python
solve(diff(func) < 0)
```

Невозможно определить интервалы убывания и возрастания функции.

```python
plot(func, (x, -100, 100))
```

```python
solve(func > 0)
```

```python
solve(func < 0)
```
